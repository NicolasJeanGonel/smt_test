```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.abs(x)**(1/2)
```

```python
x_test=np.linspace(-10,10,2000)
x_test
y_true=f(x_test)
plt.plot(x_test,y_true)
```

```python
# x_sample =np.sort(np.random.rand(100)*20-10)
x_sample=np.linspace(-10,10,50)
y_sample=f(x_sample)
plt.scatter(x_sample,y_sample)
```

```python
from smt.surrogate_models import KRG
model=KRG(theta0=[1e-2],print_prediction = False,corr="matern32")
model.set_training_values(x_sample,y_sample)
model.train()
y2=model.predict_values(x_test)
plt.figure()
plt.scatter(x_sample,y_sample,alpha=0.5,marker="x",c="orange")
plt.plot(x_test,y2)
plt.plot(x_test,y_true,alpha=0.5)
```

```python
model=KRG(theta0=[1],print_prediction = False,corr="squar_exp",eval_noise=False)
model.set_training_values(x_sample,y_sample)
model.train()
y2=model.predict_values(x_test)
plt.figure()
#plt.scatter(x_sample,y_sample)
plt.plot(x_test,y2)
plt.plot(x_test,y_true,alpha=0.5)
```
